## Business Case

In diesem Projekt analysieren wir ein umfangreiches Videospiel‑Dataset und führen verschiedene Auswertungen durch, darunter:

    1. Analyse der Genre-/Spiele‑Verteilung
        Ziel: Ermitteln, wie viele Spiele jeweils zu welchem Genre gehören, um die Gesamtverteilung im Datensatz zu verstehen.

    2. Genreverteilung pro Jahr
        Ziel: Für jedes Jahr die Verteilung der Genres bestimmen, um zeitliche Trends und Veränderungen im Genre‑Mix zu erkennen.

    3. Critic‑Scores pro Spiel und Konsole
        Ziel: Die Critic Scores der Spiele je Konsole gegenüberstellen, um Qualitätsunterschiede zwischen Plattformen sichtbar zu machen.

    4. Gesamtverkäufe pro Spiel
        Ziel: Die Gesamtverkäufe jedes Spiels über alle Konsolen hinweg ermitteln, um den kommerziellen Erfolg einzelner Titel zu bewerten.

    5. Anzahl veröffentlichter Spiele pro Jahr
        Ziel: Bestimmen, wie viele Spiele pro Jahr veröffentlicht wurden, um Aktivitätsniveaus und mögliche Marktzyklen zu erkennen.

    6. Zusammenarbeit zwischen Publishern und Developern
        Ziel: Ermitteln, wie viele Spiele pro Publisher gemeinsam mit denselben Developern veröffentlicht wurden, um Kooperationsmuster zu analysieren.

    7. Analyse der Publisher‑Anteile
        Ziel: Den Anteil der von jedem Publisher veröffentlichten Spiele bestimmen, um ihre Marktpräsenz einzuordnen.

    8. Analyse der Developer‑Anteile
        Ziel: Den Anteil der einzelnen Developer an allen veröffentlichten Spielen bestimmen, um ihre Bedeutung im Markt zu bewerten.

Diese Auswertungen ermöglichen es, Trends, Muster und Zusammenhänge im Videospielmarkt zu erkennen.
